```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# use latex for font rendering
plt.rc('text', usetex=True)

# scale ticks and labels
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)

L=3
numpoints = 20

X, Y = np.meshgrid(np.arange(-L,L, 2*L/numpoints), np.arange(-1,L, 2*L/numpoints))
U = X ** 2*(Y-X) +Y**5
V = Y**2*(Y-2*X)

# normalize arrows
N = np.sqrt(U**2 + V**2)
U = U/N
V = V/N

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

# plot a point at the origin
ax.plot(0,0, 'ro')

plt.show()
```

```python
# save the figure
fig.savefig('unstable_attractor.pdf', bbox_inches='tight')
```
